```python
import MySQLdb
import os
```

```python
usr = "showdownroot"
pwd = "pikachuate100ketchups"
database = "pokemon_profiles"
connection = MySQLdb.connect(host="localhost", user = usr, passwd = pwd, db = database)

cursor = connection.cursor()

#cursor.execute("SELECT VERSION()")
#row = cursor.fetchone()
#print("server version: ", row[0])


dir_name = "Game_Profiles/"
team_name = "trick room"
full_path = dir_name + team_name

print(full_path)

#This should be a prepared statement (cursor.execute(sql_statement, (param1, param2))) but i supply parameters so it should be fine
table_sql_statement = "CREATE TABLE IF NOT EXISTS {table_name} (pokemon_id int PRIMARY KEY AUTO_INCREMENT,ability varchar(32) DEFAULT 'UNKNOWN', item varchar(32) DEFAULT 'UNKNOWN', move1 varchar(32) DEFAULT 'UNKNOWN',move2 varchar(32) DEFAULT 'UNKNOWN',move3 varchar(32) DEFAULT 'UNKNOWN',move4 varchar(32) DEFAULT 'UNKNOWN',maxmove1 varchar(32) DEFAULT 'UNKNOWN',maxmove2 varchar(32) DEFAULT 'UNKNOWN',maxmove3 varchar(32) DEFAULT 'UNKNOWN',maxmove4 varchar(32) DEFAULT 'UNKNOWN');"


start = 0
for filename in os.listdir(full_path):
    file = open(full_path + '/' + filename, 'r')
    print(filename)
    Lines = file.readlines()
    
    for i in range(len(Lines)):
        if(Lines[i] == "--------------------\n"):
            start = i + 2
            break
        else:
            continue
    
    while(start < len(Lines)):
        #Get name
        poke_name = Lines[start].strip()
        if(poke_name != ""):
            #create table if non-existent
            print(poke_name)
            cursor.execute(table_sql_statement.format(table_name=poke_name.replace("-","_").replace("*", "X"))) 
            #print(poke_name)
        start += 1 #skip to ability section
        ability = (Lines[start].strip()).split("Ability: ")[1]####
        start += 1 #skip to item
        item = (Lines[start].strip()).split("Item: ")[1]####
        start += 2 #skip to move section
    
        cur_iter = 0
        max_iter = 8
        move_num = 1;
        maxmove_num = 1;
        move_list = []
        if(ability != "(null)"):
            move_list.append(("ability", ability))
        if(item != "(null)"):
            move_list.append(("item", item))
        
        while(cur_iter < max_iter):
            cur_iter += 1
            movei = Lines[start].strip()
            start += 1
        
            if(movei == "(null)"):
                break
            else:
                if(movei.startswith("Max") or movei.startswith("G-Max")):
                    #insert max move into table
                    col = "maxmove" + str(maxmove_num)
                    #print(col)
                    #print(movei)
                    if maxmove_num < 5:
                        move_list.append((col, movei))
                    maxmove_num += 1
                else:
                    #insert move into table
                    col = "move" + str(move_num)
                    #print(col)
                    #print(movei)
                    if move_num < 5:
                        move_list.append((col, movei))
                    move_num += 1
                           #print(move_list)
    
        #Construct format string
        #I should absolutely use prepared statements, but I am using my own data so it should be fine
        format_string = "INSERT INTO " + poke_name.replace("-","_").replace("*","X") + "("
        for i in range(len(move_list)):
            format_string += move_list[i][0]
            if(i < len(move_list) - 1):
                format_string += ","
        format_string += ") VALUES ("
        for i in range(len(move_list)):
            format_string += "\'" + move_list[i][1] + "\'"
            if(i < len(move_list) - 1):
                format_string += ","
        format_string += ");"
        #print(format_string)
    
        if(len(move_list) > 0):
            cursor.execute(format_string)
        
        start += 9 - (move_num + maxmove_num - 2)


cursor.close()
connection.close()
```
